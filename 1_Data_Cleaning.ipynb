# 1. Setting Up the Environment and Loading Data

```python
# Importing the pandas library for data manipulation and os for file operations
import pandas as pd
import os

print("Starting the simplified and robust data cleaning process...")

# Loading the original, raw dataset from the 'data/raw' folder
df = pd.read_csv('data/raw/Laptop_data.csv')

# Displaying the initial state of the data to verify it loaded correctly
print("Initial dataset shape:", df.shape)
df.head()
```

# 2. Simplified and Robust Feature Engineering

```python
# --- Step 1: Initial Cleaning ---
# Dropping any duplicate rows to ensure each entry is unique
df.drop_duplicates(inplace=True)

# Keeping only the original, essential columns needed for our core features
# This removes any potentially confusing or unnecessary data from the start
df = df[['name', 'price', 'processor', 'storage', 'display']].copy()


# --- Step 2: Engineering the 4 Most Powerful Features ---
# Engineering 'ram_gb': Extracting the numerical GB value for RAM
df['ram_gb'] = df['storage'].str.extract(r'(\d+)\s*GB').fillna(0).astype(int)

# Engineering 'ssd_gb': Extracting the numerical GB value for SSD storage
df['ssd_gb'] = df['storage'].str.extract(r'(\d+)\s*GB\s*SSD').fillna(0).astype(int)

# Engineering 'screen_size_inch': Extracting the numerical screen size in inches
df['screen_size_inch'] = df['display'].str.extract(r'(\d+\.?\d*)\"').fillna(0).astype(float)

# Engineering 'processor_brand': Simplifying the processor information into major brands
df['processor_brand'] = df['processor'].str.split().str[0].map({'Intel': 'Intel', 'AMD': 'AMD', 'Apple': 'Apple'}).fillna('Other')


print("Finished engineering the four core predictive features.")
```

# 3. Finalizing and Saving the Simplified Dataset

```python
# --- Finalization Step ---
# Selecting only our core engineered features and essential columns for the final dataset
final_columns = ['name', 'price', 'ram_gb', 'ssd_gb', 'screen_size_inch', 'processor_brand']
df_final = df[final_columns].copy()

# Performing a final quality check to remove any rows with missing or invalid data
df_final.dropna(inplace=True)
df_final = df_final[df_final['ram_gb'] > 0] # Ensuring all laptops have valid RAM information

# Saving the new, simplified, and clean dataset to the 'processed' folder
os.makedirs('data/processed', exist_ok=True)
df_final.to_csv('data/processed/cleaned_laptops.csv', index=False)

print("Simplified cleaning complete. New 'cleaned_laptops.csv' has been saved.")
print("\n--- Final Data Info ---")
df_final.info()
```
